```python
def readGraphData(file):
    f=open(file,"r")

    graph=[]
    
    vertexAndEdge=f.readline()
    v = int(vertexAndEdge.split()[0])
    e = int(vertexAndEdge.split()[1])
    #print(vertexAndEdge)
    
    for line in f.readlines():
        #print(line)
        row=[]
        row.append(int(line.split()[0]))
        row.append(int(line.split()[1]))
        row.append(int(line.split()[2]))
        graph.append(row)
    return v,e,graph


def printList(vertex,adj):
    for i in range(vertex):
        print(str(i)+"-> ",end=" ")
        print(adj[i])

def adjacencyList(vertex,graph):
    adj={}

    for i in range(vertex):
        adj[i]=[]

    for e in range(len(graph)):
        u=graph[e][0]
        v=graph[e][1]
        w=graph[e][2]
        adj[u].append((v,w))
        adj[v].append((u,w))
    return adj


def heuristic(n):
    H_dist = {
        0 : 5,
        1 : 1,
        2 : 0,
        3 : 3,
        4 : 6,
    }
    return H_dist[n]

import queue

def a_star(graph, start, goal):
    q = queue.PriorityQueue()
    q.put((heuristic(start), start, 0))
    visited = set()

    while not q.empty():
        h_cost, node, g_cost = q.get()

        if node == goal:
            return g_cost

        if node not in visited:
            visited.add(node)

            for neighbor, neighbor_cost in graph[node]:
                if neighbor not in visited:
                    f_cost = g_cost + neighbor_cost + heuristic(neighbor)
                    q.put((f_cost, neighbor, g_cost + neighbor_cost))

    return float('inf')

vertex, edge, graph = readGraphData("C:\\Users\\acer\\Desktop\\Python Lab Tasks\\Weighted-Graph.txt")
print(vertex, edge, graph)
adj = adjacencyList(vertex, graph)
print(adj)
start_node = 0
goal_node = 2
shortest_path_cost = a_star(adj, start_node, goal_node)
print(f"The shortest path cost between {start_node} and {goal_node} using A* search is {shortest_path_cost}.")
```
